```python
import json
import base64
from io import BytesIO
from PIL import Image
import os

def save_image_from_base64(base64_string, output_path):
    # Decode the base64 string to bytes
    image_data = base64.b64decode(base64_string)

    # Open as an image using PIL (Pillow)
    image = Image.open(BytesIO(image_data))

    # Save the image to a file
    image.save(output_path)

# Load your JSON file
json_file_path = "C:/Users/Acer/Documents/Fusemachines/LayoutLM/originaldataset/try.json"  # Replace with your JSON file path

output_directory = 'dataset/documents'
os.makedirs(output_directory, exist_ok=True)

# Process each line in the JSON file
with open(json_file_path, 'r', encoding='utf-8') as file:
    for line_num, line in enumerate(file, start=1):
        try:
            # Load the JSON data for this line
            json_data = json.loads(line.strip())

            # Process the JSON data as needed
            # Assuming the JSON structure is like: {"image": "base64_encoded_image_data", ...}
            base64_image = json_data['image']

            # Extract the desired portion of the file path
            last_trainees_idx = json_data['name'].rfind("\\Trainees")
            if last_trainees_idx != -1:
                start_idx = last_trainees_idx + len("\\Trainees\\")
                end_idx = json_data['name'].find("\\funsd", start_idx)

                if end_idx != -1:
                    folder_name = json_data['name'][start_idx:end_idx]
                    folder_name = folder_name.replace("\\", "_")  # Replace backslashes with underscores
                else:
                    folder_name = 'UnknownFolder'

                # Construct the output folder path and file path
    
                output_file_path = os.path.join(output_directory, f'{folder_name}.png')

                # Create the output folder if it doesn't exist
                os.makedirs(output_directory, exist_ok=True)

                # Save the image to the file
                save_image_from_base64(base64_image, output_file_path)

                print(f"Image {line_num} saved at {output_file_path}")
            else:
                print(f"Error: '\\Trainees' not found in the file path on line {line_num}")
        except json.JSONDecodeError as e:
            print(f"Error decoding JSON on line {line_num}: {e}")
```

```python
import json
import os

# Directory containing the JSON files to be merged
json_files_directory = 'C:/Users/Acer/Documents/Fusemachines/LayoutLM/dataset/trainfiles'

# Initialize an empty dictionary to store the merged JSON data
merged_json_data = {
    "qas": []
}

# Loop through JSON files in the directory
for filename in os.listdir(json_files_directory):
    if filename.endswith('.json'):
        file_path = os.path.join(json_files_directory, filename)

        try:
            # Read the JSON file
            with open(file_path, 'r', encoding='utf-8') as file:
                data = json.load(file)
            if 'qas' in data:
                for question in data['qas']:
                    # Change the file extension to ".png"
                    image_filename = os.path.splitext(filename)[0] + ".png"
                    question["image"] = image_filename
                    answer = question["answers"]
                    question["answers"] = [answer[0]['text']]
                    merged_json_data["qas"].append(question)
        except json.JSONDecodeError as e:
                print(f'Skipped file {file_path} due to JSON decoding error: {e}')

# Path to the output merged JSON file
output_json_file = 'dataset/train.json'

# Write the merged JSON data to the output file as a single JSON object
with open(output_json_file, 'w', encoding='utf-8') as output_file:
    json.dump(merged_json_data, output_file, indent=4)

print(f'Merged JSON data written to {output_json_file}')


```
